# If, Elif, Else

```python
l = [-1, 0, 1]
for i in l:
    if i > 0:
        print(f"{i} is positive")
    elif i < 0:
        print(f"{i} is negative")
    else:
        print(f"{i} is zero")
```

# For loops

```python
for i in range(5):
    print(f"i: {i}", end=", ")
print()
for i in range(2, 10, 2):
    print(f"i: {i}", end=", ")
```

# While loops, breaks, and continues

```python
while True:
    user_input = input("Enter text (or 'exit' to quit): ")
    if user_input.strip().lower() == 'exit':
        break
    if user_input.strip() == '':
        print("Empty input, please enter something.")
        continue
    print(f"You entered: {user_input}")
```

# Else in a loop

```python
for i in range(3):
    print(f"i: {i}")
else:
    print("Loop 1 completed without break.")

for i in range(3):
    if i == 1:
        print("Breaking the loop.")
        break
else:
    print("Loop 2 completed without break.")
```
