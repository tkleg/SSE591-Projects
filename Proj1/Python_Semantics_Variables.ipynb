```python
# Dynamic typing
x = 1
y = str(x) 
x = "Hello"
y += x
print(f"y is now: {y}")

# Immutable vs mutable types
a = 45
b = a
a = 5
print(f"b is still: {b}")

a = [1, 2, 3]
b = a
a.append(4)
print(f"b is: {b}")

a = [1, 2, 3]
b = a
a = [4, 5, 6]
print(f"b is still: {b}")
```
