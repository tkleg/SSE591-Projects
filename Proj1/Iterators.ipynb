# Iterator

```python
my_iter = iter([1, 2, 3, 4, 5])
while True:
    try:
        item = next(my_iter)
        print(item)
    except StopIteration:
        break
```

# Map And Filter

```python
divisible_by_5 = lambda x: x % 5 == 0
double = lambda x: x * 2
for i in filter(divisible_by_5, map(double, range(100))):
    print(i, end=' ')
```

# Enumerate

```python
for index, value in enumerate(['a', 'b', 'c']):
    print(index, value, sep=': ')

for x in enumerate(['a', 'b', 'c']):
    print(x, type(x), sep=': ')
```

# Permutations

```python
from itertools import permutations

my_set = {'Georgia', 'Alabama', 'Florida'}
for p in permutations(my_set):
    print(p)
```

# Zip

```python
list_a = [1, 2, 3]
list_b = ['a', 'b', 'c', 'd']
list_c = [10, 20, 30, 40]
for a, b, c in zip(list_a, list_b, list_c):
    print(a, b, c, sep=', ')
```
