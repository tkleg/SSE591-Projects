# Iterator

```python
my_iter = iter([1, 2, 3, 4, 5])
while True:
    try:
        item = next(my_iter)
        print(item)
    except StopIteration:
        break
```

# Map And Filter

```python
divisible_by_5 = lambda x: x % 5 == 0
double = lambda x: x * 2
for i in filter(divisible_by_5, map(double, range(100))):
    print(i, end=' ')
```

# Enumerate

```python
for index, value in enumerate(['a', 'b', 'c']):
    print(index, value, sep=': ')
```

# Permutations

```python
from itertools import permutations

my_set = {'Georgia', 'Alabama', 'Florida', 'Mississippi'}
size = 2
for p in permutations(my_set, size):
    print(p)
```
