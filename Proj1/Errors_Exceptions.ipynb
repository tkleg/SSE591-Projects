# Try-Except, Else, Finally

```python
def process_int(input_str):
    y = "original"
    try:
        y = int(input_str)
    except ValueError:
        print(f"Error: '{input_str}' is not a valid integer.")
    else:
        print(f"Successfully converted '{input_str}' to integer: {y}")
    finally:
        print(f"y = {y}")

process_int("10")
process_int("abc")
```

# Raise

```python
y = "original"
input_str = "4"
try:
    y = int(input_str)
except ValueError:
    raise ValueError(f"Error: '{input_str}' is not a valid integer.")
```
