# Basic Operators

```python
a = 5
b = 17

print("a + b =", a + b)
print("a - b =", a - b)
print("a * b =", a * b)
print("a / b =", a / b)
print("a // b =", a // b)
print("a % b =", a % b)
print("a ** b =", a ** b)
```

# Bitwise Operators

```python
print(f"a & b = {bin(a)} & {bin(b)} = {bin(a & b)}")
print(f"a | b = {bin(a)} | {bin(b)} = {bin(a | b)}")
print(f"a ^ b = {bin(a)} ^ {bin(b)} = {bin(a ^ b)}")
print(f"~a = ~{bin(a)} = {bin(~a)}")
print(f"a << b = {bin(a)} << {b} = {bin(a << b)}")
print(f"a >> b = {bin(a)} >> {b} = {bin(a >> b)}")
```

# Boolean Operators

```python
a = True
b = False
print(f"a and b = {a} and {b} = {a and b}")
print(f"a or b = {a} or {b} = {a or b}")
print(f"not a = not {a} = {not a}")
```

# Identity Operators

```python
a = 1000
b = 1000
print(f"a is b = {a is b}")
b = a
print(f"a is not b = {a is not b}")
```

# Membership Operators

```python
l = [1, 2, 3, 4, 5]
print(f"1 in l = {1 in l}")
print(f"6 not in l = {6 not in l}")
```

# Assignment Operators

```python
a = 6
print("Initial value of a:", a)
a += 3
print("After a += 3, a =", a)
a -= 2
print("After a -= 2, a =", a)
```

# Comparison Operators

```python
a = 5
b = 17
print("a == b:", a == b)
print("a != b:", a != b)
print("a > b:", a > b)
print("a < b:", a < b)
print("a >= b:", a >= b)
print("a <= b:", a <= b)
```

# Identity Operators

```python
a = 256
b = 256
print("a is b:", a is b)
a = -5
b = -5
print("a is b:", a is b)

a = 257
b = 257
print("a is b:", a is b)
a = -6
b = -6
print("a is b:", a is b)
```
