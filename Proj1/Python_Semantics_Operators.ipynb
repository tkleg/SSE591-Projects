# Basic Operators

```python
a = 5
b = 17

print("a + b =", a + b)
print("a - b =", a - b)
print("a * b =", a * b)
print("a / b =", a / b)
print("a // b =", a // b)
print("a % b =", a % b)
print("a ** b =", a ** b)
```

# Bitwise Operators

```python
print(f"a & b = {bin(a)} & {bin(b)} = {bin(a & b)}")
print(f"a | b = {bin(a)} | {bin(b)} = {bin(a | b)}")
print(f"a ^ b = {bin(a)} ^ {bin(b)} = {bin(a ^ b)}")
print(f"~a = ~{bin(a)} = {bin(~a)}")
print(f"a << b = {bin(a)} << {b} = {bin(a << b)}")
print(f"a >> b = {bin(a)} >> {b} = {bin(a >> b)}")
```

# Assignment Operators

```python
a = 6
print("Initial value of a:", a)
a += 3
print("After a += 3, a =", a)
a -= 2
print("After a -= 2, a =", a)
```

# Comparison Operators

```python
a = 5
b = 17
print("a == b:", a == b)
print("a != b:", a != b)
print("a > b:", a > b)
print("a < b:", a < b)
print("a >= b:", a >= b)
print("a <= b:", a <= b)
```

# Identity and Membership Operators

```python
a = 256
b = 256
print("a is b:", a is b)

a = 257
b = 257
print("a is b:", a is b)
```
