# Lists

```python
l = [1, 2, 3, 4, 5]
print(f"l: {l}, type: {type(l)}")
print(f"Length of l: {len(l)}")
print(f"First element: {l[0]}, Last element: {l[-1]}")
print(f"Slice l[1:4]: {l[1:4]}")
l[0] = 10
print(f"After modification, l: {l}")
l.remove(3)
print(f"After removing 3, l: {l}")
```

# Sets

```python
s = {1, 2, 3, 4, 5}
print(f"s: {s}, type: {type(s)}")
print(f"Length of s: {len(s)}")
print(f"Is 3 in s? {'Yes' if 3 in s else 'No'}")
print(f"Is 6 in s? {'Yes' if 6 in s else 'No'}")
z = {3, 4, 5, 6, 7}
print(f"Intersection of s and z: {s.intersection(z)}")
print(f"Union of s and z: {s.union(z)}")
s.remove(2)
print(f"After removing 2, s: {s}")
```

# Dictionaries

```python
d = {'a': 1, 'b': 2, 'c': 3}
print(f"d: {d}, type: {type(d)}")
print(f"Length of d: {len(d)}")
print(f"Value for key 'a': {d['a']}")
print(f"Keys of d: {list(d.keys())}")
d.update({'d': 4})
print(f"After adding key 'd', d: {d}")
d.pop('b')
print(f"After removing key 'b', d: {d}")
```

# Tuples

```python
t = (1, 2, 3, 4, 5)
print(f"t: {t}, type: {type(t)}")
print(f"Length of t: {len(t)}")
print(f"First element: {t[0]}, Last element: {t[-1]}")
print(f"Slice t[1:4]: {t[1:4]}")
try:
    t[0] = 10
except TypeError as e:
    print(e)
```
