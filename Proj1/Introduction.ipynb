```python
# Simple math problems

a = -5
b = 5
c = 0

import math

inside_of_sqrt = (b * b) - (4 * a * c)

if(inside_of_sqrt < 0):
    print("No real solutions")
else:
    sqrt_of_inside = math.sqrt(inside_of_sqrt)
    top_of_equation_1 = -b + sqrt_of_inside
    top_of_equation_2 = -b - sqrt_of_inside

    bottom_of_equation = 2 * a

    solution1 = top_of_equation_1 / bottom_of_equation
    solution2 = top_of_equation_2 / bottom_of_equation

    print("The solutions are:", solution1, "and", solution2)
```
