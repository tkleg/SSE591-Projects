# Functions

```python
def func(start, length = 10, *args, **kwargs):
    l = []
    for i in range(start, start + length):
        l.append(i)
    print(f"args: {args}")
    print(f"kwargs: {kwargs}")
    return l
```

```python
print(func(5, 5, "extra_arg1", "extra_arg2", key1="value1", key2="value2"))
print()
print(func(3, key1="value1", key2="value2"))
```
