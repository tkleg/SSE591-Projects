# Functions

```python
def func(start, length = 10, *args, **kwargs):
    l = []
    for i in range(start, start + length):
        l.append(i)
    print(f"args: {args}")
    print(f"kwargs: {kwargs}")
    return l
```

```python
print(func(5, 5, "extra_arg1", "extra_arg2", key1="value1", key2="value2"))
print()
print(func(3, key1="value1", key2="value2"))
```

```python
l = [324, 54335, 63, 234, 432]
d = {f"{i}": val for i, val in enumerate(l)}
print(f"Original dict: {d}")
sorted_d = dict(sorted(d.items(), key=lambda item: item[1]))
print(f"Sorted dict: {sorted_d}")
```
