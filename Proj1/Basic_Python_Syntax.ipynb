```python
# Comments help denote different section of code
# They also can be used to explain what a particular section of code is doing

import random, math

#Uses a semicolon to separate two statements on the same line
a = 10; b = 100

#Average the numbers for the mean of the normal distribution
mean = (a + b) / 2

#Calculates the standard deviation by finding the sum of the squared differences from the mean, 
# dividing by the number of values, and taking the square root
sum_of_squared_differences = sum([(x - mean) ** 2 for x in [a, b]])
std_dev = math.sqrt(sum_of_squared_differences / 2)

#Generates a random integer from a normal distribution.
x = int(random.normalvariate(mean, std_dev))

#Uses and f-string to interpolate the mean, standard deviation, 
# and random number into a string and prints it to the console.
#Uses a backslash to break the string into two lines for better readability.
print(f"Random number from a normal distribution with mean {mean} \
      and standard deviation {std_dev}: {x}")

#See how far the random number is from the the lower bound of the normal distribution
diff_from_a = abs(x - a)

#See how far the random number is from the the upper bound of the normal distribution
diff_from_b = abs(x - b)

#Use indentation to denote a block of code that is executed if the condition is true
#Also note that whitespace in the line here does not matter
if( diff_from_a < diff_from_b):
    print(f"{x} is closer to {a} than {b}")
else:
    print(f"{x} is closer to {b} than {a}")
```
