# Generators

```python
import sys

n = 999_999

gen_for_list = (x for x in range(n))
list_of_gen = list(gen_for_list)

gen_for_set = (x for x in range(n))
print(f"Memory Size of list_of_gen: {sys.getsizeof(list_of_gen)}")
print(f"Memory Size of gen_for_set object: {sys.getsizeof(gen_for_set)}")

set_of_gen = set(gen_for_set)
set_of_list_of_gen = set(list_of_gen)

if set_of_gen == set_of_list_of_gen:
    print("Sets are equal")
else:
    print("Sets are not equal")
```

```python
def get_gen(n):
    for x in range(n):
        print(f"Yielding: {x}")
        yield x

for x in get_gen(3):
    print(f"Received: {x}")
```
