# Numerical Types

```python
int_X = 1
float_X = 1.0
complex_X = 1 + 1j

print(f"int_X: {int_X}, type: {type(int_X)}")
print(f"float_X: {float_X}, type: {type(float_X)}")
print(f"complex_X: {complex_X}, type: {type(complex_X)}")
```

# Others

```python
bool_X = True
str_X = "Hello, World!"
print(f"bool_X: {bool_X}, type: {type(bool_X)}")
print(f"str_X: {str_X}, type: {type(str_X)}")
```

# Floating-point numbers

```python
print(0.25 + 0.75 == 1)
print(0.1 + 0.2 == 0.3)
```

# Complex

```python
c = 1 + 2j
print(f"c: {c}, type: {type(c)}")
print(f"Real part: {c.real}, Imaginary part: {c.imag}")
print(f"Conjugate: {c.conjugate()}")
print(f"abs(c): {abs(c)}")
```

# String

```python
s = "Hello, World!"
print(f"s: {s}, type: {type(s)}")
print(f"Length of s: {len(s)}")
print(f"Uppercase: {s.upper()}, Lowercase: {s.lower()}")
```

# Booleans

```python
b = True
print(f"b: {b}, type: {type(b)}")
print(f"bool of 0: {bool(0)}, bool of 1: {bool(1)}")
```
