```python
import matplotlib.pyplot as plt
import numpy as np
#plt.style.use('seaborn-v0_8-darkgrid')

def min_max_normalize(arr):
    arr = np.asarray(arr, dtype=float)
    arr_min = arr.min()
    arr_max = arr.max()
    return (arr - arr_min) / (arr_max - arr_min)
```

# subplots by hand

```python
plt.axes()
plt.axes([0.5, 0.5, 0.1, 0.1])
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.5, 0.8, 0.4])
ax2 = fig.add_axes([0.1, 0.1, 0.8, 0.4])
x = np.linspace(1, 10, 100)
ax1.plot(x, min_max_normalize(np.log(x)))
ax2.plot(x, min_max_normalize(x**2))
```

# Simple grids of subplots

```python
fig = plt.figure()
fig.subplots_adjust(hspace=0.5, wspace=0.5)
for i in range(1, 10):
    plt.subplot(3, 3, i)
    plt.text(0.5, 0.5, str((3, 3, i)), fontsize=16, ha='center')
```

# plt.subplots: the whole grid in one line

```python
fig, ax = plt.subplots(3, 3, sharex=True, sharey=True)
fig.set_size_inches(4, 4)
ax[0, 0].plot(x, x)
ax[0,1].text(5, 5, 'ax[0,1]', fontsize=12, ha='center')
```

# plt.GridSpec: more complicated arguments

```python
fig = plt.figure(figsize=(5, 4))
grid = fig.add_gridspec(3, 3, wspace=0.4, hspace=0.3)

ax1 = plt.subplot(grid[0, 0])
ax1.text(0.5, 0.5, 'ax1', fontsize=12, ha='center')

ax2 = plt.subplot(grid[0, 1])
ax2.text(0.5, 0.5, 'ax2', fontsize=12, ha='center')

ax3 = plt.subplot(grid[:,2])
ax3.text(0.5, 0.5, 'ax3', fontsize=12, ha='center')

ax4 = plt.subplot(grid[1:, :2])
ax4.text(0.5, 0.5, 'ax4', fontsize=12, ha='center')
```
