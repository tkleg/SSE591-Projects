```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

def min_max_normalize(arr):
    arr = np.asarray(arr, dtype=float)
    arr_min = arr.min()
    arr_max = arr.max()
    return (arr - arr_min) / (arr_max - arr_min)
def max_abs_normalize(arr):
    arr = np.asarray(arr, dtype=float)
    arr_max_abs = np.max(np.abs(arr))
    return arr / arr_max_abs
```

# min-max normalization

```python
x = np.linspace(0, 10, 100)
y1 = 3 * np.sin(x)
y2 = np.exp(x)

fig, ax = plt.subplots(3, 3, figsize=(11, 8))

row_defs = [
    ("Original", lambda y: y),
    ("Min-Max", min_max_normalize),
    ("Max-Abs", max_abs_normalize),
]

col_defs = [
    (" 3*sin(x)", [("3*sin(x)", y1)]),
    ("exp(x)", [("exp(x)", y2)]),
    ("3*sin(x) + exp(x)", [("3*sin(x)", y1), ("exp(x)", y2)]),
]

for row, (row_label, func) in enumerate(row_defs):
    for col, (col_label, series_list) in enumerate(col_defs):
        for series_label, y in series_list:
            ax[row, col].plot(x, func(y), label=series_label)

        # Top edge labels (column titles)
        if row == 0:
            ax[row, col].set_title(col_label)

        # Left edge labels (row titles)
        if col == 0:
            ax[row, col].set_ylabel(row_label)

        ax[row, col].legend()

plt.tight_layout()
```
