```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn-v0_8-darkgrid')
```

# Visualizing a 3d Function

```python
def f(x, y):
    return np.cos( x * np.sqrt( np.abs( np.log( np.abs(y + 4)) ) ) ) * np.sin( np.log(np.abs(y)) * np.sqrt(np.abs(x - 4)) )
```

```python
x = np.linspace(-10, 10, 1000)
y = np.linspace(-10, 10, 1000)

levels = 6
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
plt.figure(figsize=(6, 4))
plt.contour(X, Y, Z, levels, cmap='RdGy')
plt.colorbar()
```

```python
contours = plt.contour(X, Y, Z, levels, cmap='RdGy')
plt.clabel(contours, inline=True, fontsize=8)
plt.colorbar()
```

```python

plt.contourf(X, Y, Z, levels, cmap='RdGy')
plt.colorbar()
#contourf fills the area between contour lines with color, while contour only draws the lines.
```

```python
plt.imshow(Z, extent=(-10, 10, -10, 10), cmap='RdGy')
plt.colorbar()
# no x and y grid, specify the range on both axis.
```
