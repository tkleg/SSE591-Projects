```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn-v0_8-darkgrid')
```

```python
x = np.linspace(-1, 1, 300)
y = x * np.sin( 1 / x )
plt.figure(figsize=(4, 3))
plt.plot(x, y)
```

# Adjusting the plot: line colors and styles

```python
plt.figure(figsize=(4, 3))
plt.plot(x, y, color='red', linestyle='--', label='y = x * sin(1/x)')
plt.plot(x, x * 2, color=(0, 0, 1), linestyle=':', label='y = 2x')
plt.plot(x, x * 3, color='#00FF00', linestyle='-', label='y = 3x')
plt.plot(x, x * 4, color='0.5', linestyle='dotted', label='y = 4x')
plt.plot(x, x * 5, '-b', label='y = 5x')
plt.legend()
```

# Adjusting the plot: axes limits

```python
x = np.linspace(-10, 10, 1000)
plt.figure(figsize=(4, 3))
plt.plot(x, np.tan(x))
plt.xlim(-15, 15)
plt.ylim(-100, 100)
```

```python
plt.plot(x, np.log(x))
plt.axis([-10, 10, -5, 5])
```

# Labeling plots
label the axes and add a title to the plot

```python
x = np.linspace(-10, 10, 100_000)
plt.figure(figsize=(4, 3))
plt.plot(x, np.log( np.sin( x ) ), label ='log(sin(x))')
plt.title('Plot of log(sin(x))')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
```

```python
x = np.linspace(0, 10, 100_000)
plt.plot(x, x, ':r', label='linear' )
plt.plot(x, np.repeat(1, x.shape), '.b', label='constant')
plt.plot(x, np.log(x), '-g', label='logarithmic')
plt.legend()

plt.xlim(-0.5, 10.5)
plt.ylim(-0.5, 10.5)
```
