```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

# 3d points and lines

```python
fig = plt.figure(figsize=(5, 4))
ax = fig.add_subplot(111, projection='3d')

z = np.linspace(-5, 5, 100)
x = np.cos(z)
y = np.log(np.abs(z))
ax.plot3D(x, y, z, 'gray')
z_points = np.random.uniform(-5, 5, size=20)
x_points = np.cos(z_points)
y_points = np.log(np.abs(z_points))
ax.scatter3D(x_points, y_points, z_points, c=z_points, cmap='viridis')
ax.set_xlabel('x', fontsize=14)
ax.set_ylabel('y', fontsize=14)
ax.set_zlabel('z', fontsize=14)
```

# 3d contour plots

```python
def f(x, y):
    return np.cos( np.pow( np.abs(x), 1.5 ) ) * np.log( np.abs(y) )

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
levels = 50
fig = plt.figure(figsize=(5, 4))
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, levels, cmap='binary')
ax.set_xlabel('x', fontsize=14)
ax.set_ylabel('y', fontsize=14)
ax.set_zlabel('z', fontsize=14)
ax.set_title('Contour', fontsize=16)
```

# wireframes and surface plots
easy visualization of 3d data

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_wireframe(X, Y, Z, color='blue', rstride=6, cstride=6)
ax.set_xlabel('x', fontsize=14)
ax.set_ylabel('y', fontsize=14)
ax.set_zlabel('z', fontsize=14)
ax.set_title('Wireframe', fontsize=16)
```

```python
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=5, cstride=5, cmap='viridis')
ax.set_xlabel('x', fontsize=14)
ax.set_ylabel('y', fontsize=14)
ax.set_zlabel('z', fontsize=14)
ax.set_title('Surface', fontsize=16)
```

# surface triangulations

```python
def g(x, y):
    return np.cos( np.pow( np.abs(x), 0.25 ) * y )
x = np.linspace(-5, 5, 25)
y = np.linspace(-5, 5, 25)
X, Y = np.meshgrid(x, y)
Z = g(X, Y)
ax = plt.figure(figsize=(5, 4)).add_subplot(111, projection='3d')
ax.plot_trisurf(X.flatten(), Y.flatten(), Z.flatten(), cmap='plasma')
ax.set_xlabel('x', fontsize=14)
ax.set_ylabel('y', fontsize=14)
ax.set_zlabel('z', fontsize=14)
ax.set_title('Surface', fontsize=16)
```
