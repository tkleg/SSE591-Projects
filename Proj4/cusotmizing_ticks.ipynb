```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

# major and minor ticks

```python
fig, ax = plt.subplots()
fig.set_size_inches(5, 3)
ax.set_xscale('log')
ax.set_yscale('log')
ax.grid()
```

# hide ticks and labels

```python
ax = plt.axes()
ax.yaxis.set_major_locator(plt.NullLocator())
ax.xaxis.set_major_locator(plt.NullLocator())
x = np.linspace(1, 1000, 1000)
y = np.cumsum(np.random.randn(1000))
plt.plot(x, y)
```

```python
fig, ax = plt.subplots(3, 3)
for i in range(3):
    for j in range(3):
        ax[i, j].plot(x, np.sin(x ** 2))
        ax[i, j].xaxis.set_major_locator(plt.NullLocator())
        ax[i, j].yaxis.set_major_locator(plt.NullLocator())
```

# reducin or increasing the number of ticks

```python
fig, ax = plt.subplots(3, 3, sharex=True, sharey=True)
for i in range(3):
    for j in range(3):
        ax[i, j].plot(x, np.sin(x * (i + j)))
        ax[i, j].xaxis.set_major_locator(plt.MaxNLocator(4))
        ax[i, j].yaxis.set_major_locator(plt.MaxNLocator(4))
```

# fancy tick formats

```python
fig, ax = plt.subplots()
x = np.linspace(0, 20, 1000)
ax.plot(x, np.cos(x))
ax.grid(True)
ax.xaxis.set_major_locator(plt.MultipleLocator(np.pi))
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda val, pos: f'{val/np.pi:.0f}π' if val != 0 else '0'))
```

```python
def my_formatter(x, pos):
    return f"{x}:{pos}"
fig, ax = plt.subplots()
x = np.linspace(0, 20, 1000)
ax.plot(x, np.cos(x))
ax.grid(True)
ax.xaxis.set_major_formatter(plt.FuncFormatter(my_formatter))
```
