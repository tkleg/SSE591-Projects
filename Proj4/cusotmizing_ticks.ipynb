# hide ticks and labels

```python
ax = plt.axes()
ax.yaxis.set_major_locator(plt.NullLocator())
ax.xaxis.set_major_locator(plt.NullLocator())
x = np.linspace(0, 5, 1000)
indices_to_remove = np.random.choice(len(x), size=940, replace=False)
x = np.delete(x, indices_to_remove)
plt.plot(x, np.sin( x** 2 ))
```

```python
fig, ax = plt.subplots(3, 3)
for i in range(3):
    for j in range(3):
        ax[i, j].plot(x, np.sin(x ** 2))
        ax[i, j].xaxis.set_major_locator(plt.NullLocator())
        ax[i, j].yaxis.set_major_locator(plt.NullLocator())
```

# reducin or increasing the number of ticks

```python
fig, ax = plt.subplots(3, 3, sharex=True, sharey=True)
for i in range(3):
    for j in range(3):
        ax[i, j].plot(x, np.sin(x ** 2))
        ax[i, j].xaxis.set_major_locator(plt.MaxNLocator(4))
        ax[i, j].yaxis.set_major_locator(plt.MaxNLocator(4))
```

# fancy tick formats

```python
fig, ax = plt.subplots()
x = np.linspace(0, 20, 1000)
ax.plot(x, np.cos(x))
ax.grid(True)
ax.xaxis.set_major_locator(plt.MultipleLocator(np.pi))
```

```python
def my_formatter(x, pos):
    return f"{x}:{pos}"
fig, ax = plt.subplots()
x = np.linspace(0, 20, 1000)
ax.plot(x, np.cos(x))
ax.grid(True)
ax.xaxis.set_major_formatter(plt.FuncFormatter(my_formatter))
```
