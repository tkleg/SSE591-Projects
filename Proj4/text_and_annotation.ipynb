```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

plt.style.use('seaborn-v0_8-darkgrid')
diamonds = sns.load_dataset("diamonds")
```

```python
diamonds.info()
```

```python
diamonds = diamonds.query('color == "D" & cut == "Ideal"')
diamonds.sort_values('carat', inplace=True)
plt.plot(diamonds.carat, diamonds.price)
```

```python
fig, ax = plt.subplots()
ax.axis([0, 5, 0, 5])
ax.text(0.5, 0.5, "Hello World", fontsize=20, color='red', alpha=0.5, rotation=45,
        transform=ax.transAxes, ha='center', va='center')
ax.text(3, 3, 'Max Price', transform=ax.transData)
ax.text(0, 0, 'Max Price', transform=fig.transFigure)
```

# arrows and annotations

```python
x = np.linspace(-5, 5, 100)
y = x**2 - 3*x + 5
fig, ax = plt.subplots()
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 40)
ax.plot(x, y)
ax.annotate('local min', xy=(1.5, 2.75), xytext=(2.5, 20), arrowprops=dict(facecolor='black', connectionstyle='arc3, rad=0.5'))
```

```python
x = np.linspace(-5, 5, 100)
y = x**2 - 3*x + 5
fig, ax = plt.subplots()
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 40)
ax.plot(x, y)
ax.annotate('local min', xy=(1.5, 2.75), xytext=(2.5, 20), 
    arrowprops=dict(facecolor='black', connectionstyle='arc3, rad=0.5')
)
ax.annotate('origin', xy=(0, 0), xytext=(-2, 0), transform=ax.transData, fontsize=12, 
    color='blue', alpha=0.5, arrowprops=dict(arrowstyle='->', color='blue', connectionstyle='arc3, rad=0.5')
)
plt.title('x^2 - 3x + 5')
plt.xlabel('x')
plt.ylabel('y')
```
