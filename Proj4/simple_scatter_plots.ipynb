```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

plt.style.use('seaborn-v0_8-darkgrid')
```

# Scatter plots with plt.plot

```python
x = np.linspace(-10, 10, 75)
plt.plot(x, np.log(x), 'o')
```

```python
plt.plot(x, np.log(x), ':o')
```

# scatter plots with plt.scatter

```python
plt.scatter(x, np.log(x))
```

```python
x = np.random.rand(100)
y = np.random.rand(100)
colors = np.random.rand(100)
sizes = np.random.rand(100)

plt.scatter(x, y, c=colors, s=sizes*1000, alpha=0.5, cmap='viridis')
plt.colorbar()
```

```python
taxis = sns.load_dataset('taxis')
taxis.info()
```

```python
plt.xlabel('Distance')
plt.ylabel('Total')
plt.scatter(taxis.distance, taxis.total, c=taxis.color, s= 30 * taxis.passengers, alpha=0.5)
```

plt.scatter renders each point seperately, which takes more time. plt.plt renders the points as copies, so it is faster. However, plt.scatter allows for more customization of the points, such as color and size.
