```python
import pandas as pd
import numpy as np
import time
```

# Tradeoffs in handling missing values
masking: bool array indicating which values are missing.
- allows for efficient filtering and processing of data
- requires additional storage for the bool array
  
sentinel value: using a specific value like NaN or "" to represent missing data.
- prevents a value from being used as real data
- extra time and effort to process
- this is what pandas uses, they wanted to avoid overhead. NaN and None.

```python
df = pd.DataFrame(columns=['Type', 'Time (ms)'])
for type in ['object', 'int32', 'int64', 'float32', 'float64']:
    start = time.time() * 1000
    np.sort(np.random.rand(1_000_000).astype(type))
    end = time.time() * 1000
    df.loc[len(df)] = {'Type': type, 'Time (ms)': end - start}
df.sort_values(by='Time (ms)', ascending=True)
```

## nan in numpy

```python
arr = np.array([1, 2, 3, np.nan, 5], dtype=float)
print(np.mean(arr))
print(np.nanmean(arr))
```

## nan and none in pandas

```python
series = pd.Series([1, 2, 3, None, 5])
print(series)
```

if the type has no sentinel value, the series will be upcasted to a type with said sentinel. For example, if you have a series of integers and you introduce a None value, the series will be upcasted to a float type to accommodate the None value.

# operating on null values
- isnull() returns a boolean series indicating which values are null (NaN or None).
- notnull() returns a boolean series indicating which values are not null.
- dropna() returns a series with the null values removed.
- fillna() returns a series with the null values filled in with a specified value.

```python
df = pd.DataFrame({'A': [None, None, 3, 6, 5], 'B': [None, 2.0, np.nan, 4.0, 5.0]})
mask = df.notnull()
print(mask, "\n")
print(df.dropna())
```

## filling null

```python
cols = ['A', 'B', 'C', 'D', 'E']
dictionary = {}
for col in cols:
    dictionary[col] = [np.nan if np.random.rand() > 0.6 else x for x in np.random.randint(0, 100, size=3)]
df = pd.DataFrame(dictionary)
```

```python
print(df, "\n")
print(df.fillna(0), "\n")
print(df.ffill(), "\n")
print(df.bfill(), "\n")
```
