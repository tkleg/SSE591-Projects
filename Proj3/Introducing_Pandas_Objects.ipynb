```python
import pandas as pd
import numpy as np
```

# Introducing Pandas Objects

# Series
1d array with indexing

```python
series = pd.Series([1, 3, 5, 35, 6, 8])
print(series)
print(series.values)
print(series.index)
```

## special dictionary

```python
print('series[0]:\n', series[0], '\n')
print('series[1:4]:\n', series[1:4], '\n')
print('series[[1, 3, 5]]:\n', series[[1, 3, 5]], '\n')
print('series[series > 5]:\n', series[series > 5], '\n')
print('series[::-1]:\n', series[::-1], '\n')
```

## Constructing
like a numpy array except you can use noninteger indices. You can use a dictionary to construct a series. The keys of the dictionary become the indices of the series, and the values become the data.

```python
series = pd.Series([32, 234, 62, 6432, 2346], index=['index1', 'index2', 'index3', 'index4', 'index5'])
print(series, '\n')
print(series['index1'], '\n')
```

```python
mydict = {'index1': 32, 'index2': 234, 'index3': 62, 'index4': 6432, 'index5': 2346}
series = pd.Series(mydict)
print(series, '\n')
print(series['index1':'index3'], '\n')
try:
    print(series[0], '\n')
except KeyError as e:
    print(f"KeyError: {e}")
```

```python
data = {2: 32, 4: 234, 6: 62, 8: 6432, 10: 2346}
series = pd.Series(data, index=[6,2])
print(series, '\n')
```

# The Pandas DataFrame Object
speical 2d array with indexing. like a list of series. 2d numpy array.

```python
densities = pd.Series({'Ag': 10.5, 'Au': 19.3, 'Pb': 11.3, 'Pt': 21.4, 'Sn': 7.3})
electrons = pd.Series({'Ag': 47, 'Au': 79, 'Pb': 82, 'Pt': 78, 'Sn': 50})
df = pd.DataFrame({'density': densities, 'electrons': electrons})
print(df, '\n')
print(df.index, '\n')
print(df.columns, '\n')
```

# special dict
like a dict in the sense that brackets returns a column, whereas a numpy 2d array returns a row.

```python
series = df['density']
print(type(series), '\n')
print(series, '\n')
```

## constructing
collection of series. can be made from series

### single series

```python
seriesA = pd.Series({'Washington': 7.3, 'Oregon': 4.2, 'California': 38.8})
df = pd.DataFrame(seriesA, columns=['GDP'])
print(df, '\n')
```

### list of dicts
note the nan for missing values

```python
data = pd.DataFrame([{'a': 1, 'b': 2}, {'a': 5, 'b': 10, 'c': 20}])
print(data, '\n')
```

### multiple series

```python
capitals = pd.Series({'Mexico': 'Mexico City', 'Canada': 'Ottawa', 'USA': 'Washington'})
population = pd.Series({'Mexico': 122_000_000, 'Canada': 37_000_000, 'USA': 318_000_000})
df = pd.DataFrame({'capital': capitals, 'population': population})
print(df)
```

### 2d numpy array

```python
data = pd.DataFrame(np.random.randint(0, 100, size=(5, 4)), columns=list('ABCD'), index=list('abcde'))
print(data)
```

### numpy structured array

```python
data = pd.DataFrame(np.zeros(5, dtype=[('A', 'i4'), ('B', 'f4'), ('C', 'c16')]))
print(data, '\n')
```

# index object
like an array of labels. immutable, which helps maintain data integrity. can contain duplicates. 

```python
index = pd.Index([2, 3, 5, 7, 11])
print(index, '\n')
print(index[1], '\n')
print(index[1:4], '\n')
```

## like a set

```python
index1 = pd.Index([2, 3, 5, 7, 11])
index2 = pd.Index([1, 3, 5, 7, 9])
print(index1 & index2, '\n')
print(index1 | index2, '\n')
print(index1 ^ index2, '\n')
```
