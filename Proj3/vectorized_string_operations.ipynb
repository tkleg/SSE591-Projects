````python
import pandas as pd
import html
class display(object):
    """Display HTML representation of multiple objects"""
    template = """<div style="float: left; padding: 10px;">
    <p style='font-family:"Courier New", Courier, monospace'>{0}</p>{1}
    </div>"""
    def __init__(self, *args):
        self.args = args
        
    def _html_for(self, obj):
        if hasattr(obj, '_repr_html_'):
            try:
                rendered = obj._repr_html_()
                if rendered is not None:
                    return rendered
            except Exception:
                pass
        return '<pre>{}</pre>'.format(html.escape(repr(obj)))
        
    def _repr_html_(self):
        return '\n'.join(self.template.format(a, self._html_for(eval(a)))
                         for a in self.args)
    
    def __repr__(self):
        return '\n\n'.join(a + '\n' + repr(eval(a))
                           for a in self.args)

def make_df(cols, ind):
    data = {c: [str(c) + str(i) for i in ind]
            for c in cols}
    return pd.DataFrame(data, ind)
````

# introducing pandas string operations

## Regex

````python
streets.str.extract(r'(.*) ').fillna("not found")
display('streets', 'streets.str.extract(r"(.*) ").fillna("not found")')
````
