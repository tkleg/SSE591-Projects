```python
import pandas as pd
import seaborn as sns
```

```python
taxis = sns.load_dataset('taxis')
taxis.info()
```

# HistPlot

```python
sns.histplot(data=taxis, x='fare', hue='payment', multiple='stack', bins = 30)
```

# Lineplot

```python
sns.lineplot(data=taxis, x='fare', y='total', hue='payment')
```

# Boxplot

```python
sns.boxplot(data=taxis, x='pickup_borough', y='distance', showfliers=False)
```

# Scatter

```python
sns.scatterplot(data=taxis, x='distance', y='fare', hue='payment', alpha=0.5)
```
