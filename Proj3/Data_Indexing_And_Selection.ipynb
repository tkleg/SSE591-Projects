```python
import pandas as pd
import numpy as np
```

# data selection in series

```python
data = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e'])
print(data['a'], '\n')
print('a' in data, '\n')
print(data.keys(), '\n')
print(list(data.items()), '\n')
data['e'] = 6
print(data['b':'d'])
print(data[0:2])
print(data)
```

## 1d array with slice

## masking and fancy indexing

```python
print(data[data > 2])
print(data[['a', 'b', 'c']])
```

## fancy indexing

## indexers: loc, iloc, ix
loc uses labels, iloc uses integer positions, ix is a hybrid of the two
ix is removed in current pandas versions

```python
numdata = pd.Series(np.arange(10)[::-1], index=list('abcdefghij'))
print(numdata.loc['a'], '\n')
print(numdata.loc['c':'g'], '\n')
print(numdata.iloc[3:6], '\n')
```

# Data selection in DataFrame
like a structured array.

```python
maker = pd.Series(['Toyota', 'Honda', 'Ford'])
model = pd.Series({'Toyota': 'Camry', 'Honda': 'Civic', 'Ford': 'F-150'})
weight = pd.Series({'Toyota': 1500, 'Honda': 1200, 'Ford': 4000})
cost = pd.Series({'Toyota': 20000, 'Honda': 22000, 'Ford': 30000})
cars = pd.DataFrame({'model': model, 'weight': weight, 'cost': cost})
print(cars)
```

```python
print(cars['model'], '\n')
print(cars.model, '\n')
```

## making new columns

```python
cars['costPerPound'] = cars['cost'] / cars['weight']
print(cars)
```

## 2d array
uses values

```python
print(cars.values, '\n')
print(cars.values[0], '\n')
```

## swap rows and columns

```python
print(cars, '\n')
print(cars.T)
```

## access

```python
print(cars.iloc[1:2, :2], '\n')
print(cars.loc['Toyota':"Honda", 'costPerPound'], '\n')
print(cars.loc[cars.weight > 2000, 'model'])
```

## editing with iloc and loc

```python
print(cars, '\n')
cars.iloc[0,1] = 0
cars.loc['Honda', 'model'] = "Odyssey"
print(cars)
```

## filtering with masking

```python
print(cars[cars.costPerPound > 10], '\n')
print(cars[(cars.costPerPound > 10) & (cars.weight > 1000)], '\n')
```
