```python
import pandas as pd
import numpy as np
```

# Ufuncs index preservation

```python
seriesA = pd.Series(np.arange(1, 5))
print( np.log(seriesA), '\n' )
seriesB = pd.Series(np.arange(11, 15))
print( np.log(seriesB), '\n' )
df = pd.DataFrame({"A": seriesA, "B": seriesB})
print( np.log(df) )
```

# Adding series index preservation

```python
seriesA = pd.Series(np.arange(1, 11), index=list("abcdefghij"))
seriesA.iloc[0] = np.nan
seriesB = pd.Series(np.arange(30, 20, -1), index=list("jihgfedcba"))
print( seriesA + seriesB, '\n')
```

```python
df1 = pd.DataFrame(np.arange(9).reshape(3, 3), index=list("abc"), columns=list("ABC"))
df2 = pd.DataFrame(np.arange(12).reshape(4, 3), index=list("abcd"), columns=list("CBA"))
print( df1 + df2 )
```

# UFuncs: operating between dataframe and series

```python
df = pd.DataFrame(np.arange(12).reshape((3, 4)), index=list("abc"), columns=list("ABCD"))
print(df, '\n')
print(df.add(df['A'], axis=0), '\n') # axis = 0 means add the series to each column of the DataFrame, aligning on index
seriesA = pd.Series({'A': 9, 'C': 9})
print(df.add(seriesA, axis=1), '\n') # axis = 1 means add the series to each row of the DataFrame, aligning on columns
```
