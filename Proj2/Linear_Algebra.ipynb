```python
import numpy as np
```

```python
mat = np.array([[1, 2], [3, 4]])
print("inverse", np.linalg.inv(mat), sep="\n")
```

```python
mat = np.random.randint(1, 10, size=(9, 9))
print("determinant", np.linalg.det(mat), sep="\n")
```

```python
mat = np.array([[1, 2], [3, 4]])
print("transpose", np.transpose(mat), sep="\n")
```

```python
mat = np.random.randint(1, 10, size=(6, 6))
b = np.random.randint(1, 10, size=6)
x = np.linalg.solve(mat, b)
print("mat", mat, sep="\n")
print("b", b, sep="\n")
print("solution", x, sep="\n")
```
