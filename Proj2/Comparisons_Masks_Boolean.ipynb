```python
import numpy as np
```

```python
arr = np.array([1, 2, 3, 4, 5])
arr_bool = arr > 3
print(arr_bool)
print(np.count_nonzero(arr_bool))
arr_bool = arr < 3
print(arr_bool)
print(np.count_nonzero(arr_bool))
```

```python
arr = np.arange(11)
print(np.any(arr > 5))
print(np.all(arr < 10))
```

```python
arr = np.arange(100)
in_range = (arr > 50) & (arr < 75)
print(np.sum(in_range))
print(np.sum(~in_range))
```

```python
arr = np.random.random(100)
print(np.sum(arr > 0.5))
print( np.mean(arr[arr > 0.5]) )
```
