# Overhead

```python
import sys
import pandas as pd
import numpy as np

x = 5
print(x, type(x), sys.getsizeof(x))

# Explicit dtypes: Length as int, size columns as float
df = pd.DataFrame(columns=['Length', 'List Size Per Element', 'Array Size Per Element', 'Numpy Array Size Per Element'])

import array
for length in range(1, 200):
    l = [0] * length
    arr = array.array('i', l)
    numpy_arr = np.array(l, dtype=np.int32)
    list_size = sys.getsizeof(l) / length
    array_size = sys.getsizeof(arr) / length
    numpy_size = sys.getsizeof(numpy_arr) / length
    df.loc[len(df)] = {
        'Length': int(length),
        'List Size Per Element': list_size,
        'Array Size Per Element': array_size,
        'Numpy Array Size Per Element': numpy_size
    }

import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.plot(df['Length'], df['List Size Per Element'], label='List Size Per Element')
plt.plot(df['Length'], df['Array Size Per Element'], label='Array Size Per Element')
plt.plot(df['Length'], df['Numpy Array Size Per Element'], label='Numpy Array Size Per Element')
plt.title('Average Size of List vs Array by Length')
plt.xlabel('Length')
plt.ylabel('Average Size (bytes)')
plt.legend()
plt.show()
print(df.iloc[-1])
```
