```python
import numpy as np
```

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr
```

```python
import time

arr = np.random.randn(5_000)
arr_copy = arr.copy()
start_time = time.time()
bubble_sort(arr)
end_time = time.time()
print("Time taken to sort the array with bubble sort:", end_time - start_time, "seconds")

start_time = time.time()
np.sort(arr_copy)
end_time = time.time()
print("Time taken to sort the array with NumPy's sort:", end_time - start_time, "seconds")
```

```python
arr = np.array([3, 1, 4, 1, 5, 9, 2, 6, 5])
sorted_indices = np.argsort(arr)
print("Original array:", arr)
print("Indices that would sort the array:", sorted_indices)
```

```python
arr = np.random.randint(1, 100, size=(5, 5))
print("Original array:\n", arr)
sorted_arr = np.sort(arr, axis=0)
print("Array sorted along the first axis (columns):\n", sorted_arr)
sorted_arr = np.sort(arr, axis=1)
print("Array sorted along the last axis (rows):\n", sorted_arr)
```
