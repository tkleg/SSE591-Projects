```python
import numpy as np
np.random.seed(0)
```

```python
indeces = np.array([0, 2, 4])
arr = np.array([10, 20, 30, 40, 50])
print("arr[indeces]:", arr[indeces])
```

```python
arr = np.random.randint(1, 100, size=(5, 5))
indicesRow = np.random.randint(0, 5, size=3)
indicesCol = np.random.randint(0, 5, size=3)
print("arr:\n", arr)
print("indicesRow:", indicesRow)
print("indicesCol:", indicesCol)
print("arr[indicesRow, indicesCol]:", arr[indicesRow, indicesCol])
```

```python
arr[indicesRow[:, np.newaxis], indicesCol]
```

```python
arr = np.random.randint(1, 100, size=(1000, 1000))
row = np.random.randint(0, 1000, size=5)
col = np.random.randint(0, 1000, size=5)
print("arr[row, col]:", arr[row, col])
```

```python
# Broadcasting

arr = np.arange(1, 1001).reshape(10, 100)
row = np.random.randint(0, 10, size=5)
col = np.random.randint(0, 100, size=5)
print("arr[row[:, np.newaxis], col]:", arr[row[:, np.newaxis], col])
```

```python
# Fancy and simple indexing
arr = np.arange(1, 101).reshape(10, 10)
col = np.random.randint(0, 10, size=5)
print("arr[0, col]:", arr[0, col])

# Fancy and slicing
print("arr[[0,1], 2:5]:", arr[[0, 1], 2:5])

# Fancy and masking
selected_rows = arr[[0, 1], :]
mask = selected_rows > 6
print("selected_rows[mask]:", selected_rows[mask])
```

```python
arr = np.random.randint(1, 100, size=(10, 10))
rand = np.random.multivariate_normal(mean=np.zeros(10), cov=np.eye(10), size=10)
print("arr[rand > 0]:", arr[rand > 0])
```

```python
arr = np.random.randint(1, 100, size=1000)

import matplotlib.pyplot as plt

indices = np.random.choice(np.arange(len(arr)), size=50, replace=False)
selected_elements = arr[indices]
plt.scatter(np.arange(len(arr)), arr, s=12, alpha=0.6)
plt.scatter(indices, selected_elements, facecolors='none', edgecolors='red', s=70, linewidths=1.5)
plt.show()
```

```python
arr = np.arange(10)
indices = np.array([2, 5, 9])
arr[indices] *= 9999
print(arr)
```

```python
arr = np.random.randint(1, 100, size=1000)
bins = np.linspace(1, 100, 25)
counts = np.zeros_like(bins)
i = np.searchsorted(bins, arr)
np.add.at(counts, i, 10)

plt.hist(arr.flatten(), bins=bins, edgecolor='black')
```
