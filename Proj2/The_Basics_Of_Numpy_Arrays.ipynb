# NumPy Array Attributes

```python
import numpy as np
mat = np.random.randint(1, 100, size=(4, 4))
mat
```

```python
print("ndim:", mat.ndim)
print("shape:", mat.shape)
print("size:", mat.size)
print("dtype:", mat.dtype)
print("itemsize:", mat.itemsize)
print("nbytes:", mat.nbytes)
```

```python
print(mat, end="\n\n")
print(mat[0], end="\n\n")
print(mat[-2], end="\n\n")
```

```python
print(mat, end="\n\n")
print(mat[0,1], end="\n\n")
print(mat[1:3, 2:3], end="\n\n")
print(mat[2:4], end="\n\n")
print(mat[::2, ::2], end="\n\n")
```

```python
print(mat[:, 1], end="\n\n")
print(mat[1, :], end="\n\n")
```

```python
print(mat, end="\n\n")
submat = mat[1:3, 1:3]
submat[:] = 999
print(mat, end="\n\n")
```

```python
mat_copy = mat.copy()
mat_copy[0, :] = 111
print(mat, end="\n\n")
```
