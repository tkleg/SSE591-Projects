```python
import numpy as np
import time

arrA = np.random.randint(1, 100, size=(1000, 1000))
arrB = arrA.copy()

start_time_loop = time.time()
for i in range(1000):
    for j in range(1000):
        arrA[i, j] += 1
end_time_loop = time.time()
print("Time taken:", end_time_loop - start_time_loop, "seconds")

start_time_numpy = time.time()
arrB += 1
end_time_numpy = time.time()
print("Time taken:", end_time_numpy - start_time_numpy, "seconds")

print("Are the two arrays equal?", np.array_equal(arrA, arrB))
```

```python
mat = np.random.randint(1, 10, size=(4, 4))
print(mat)
print(mat % 2)
```

```python
mat2 = np.random.randint(1, 10, size=(4, 4))
print(mat, mat2, mat + mat2, sep="\n\n")
```

```python
mat = np.random.randint(-6, 10, size=(3, 3))
print(mat)
print(3 * mat - 1)
print(np.abs(mat))
```

```python
mat = np.random.randint(1, 100, size=(3, 3))
print(mat, "\n")
print("sin", np.sin(mat), "\n")
print("cos", np.cos(mat))
```

```python
mat = np.random.randint(1, 10, size=(3, 3))
print(mat, "\n")
print("exp", np.exp(mat), "\n")
print("log", np.log(mat), "\n")
print("exp - 1", np.expm1(mat), "\n")
print("log1p", np.log1p(mat), "\n")
```

```python
from scipy import special

mat = np.random.randint(1, 10, size=(3, 3))
print("erf", special.erf(mat))
```
