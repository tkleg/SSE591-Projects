```python
import numpy as np
```

```python
arr1 = np.arange(5)
arr2 = np.arange(10, 20).reshape(2, 5)

print(arr1, "\n")
print(arr2, "\n")
print(arr2 + arr1, "\n")
```

```python
arr1 = np.arange(5)
arr2 = np.arange(5).reshape(5, 1)
arr1 + arr2
```

```python
arr1 = np.arange(100).reshape(1, 100)
arr2 = np.arange(5).reshape(5, 1)

print("arr1 shape", arr1.shape, "\n")
print("arr2 shape", arr2.shape, "\n")
print("arr1 + arr2 shape", (arr1 + arr2).shape, "\n")
```

```python
arr1 = np.arange(5)
arr2 = np.arange(20).reshape(2, 10)
arr1 + arr2
```

```python
arr = np.random.randint(1, 100, size=(4, 4))
arr_avg = np.mean(arr, axis=0)
arr_centered = arr - arr_avg
print("arr\n", arr, "\n")
print("arr_avg\n", arr_avg, "\n")
print("arr_centered\n", arr_centered, "\n")
```

```python
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = x.reshape(1, 100)
z = np.cos(x+y) + np.sin(y*x) * 5.5

plt.imshow(z, extent=[0, 10, 0, 10], origin='lower', cmap='viridis')
plt.colorbar()
```
