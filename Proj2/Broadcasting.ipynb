```python
arr = np.random.randint(1, 100, size=(4, 4))
arr_avg = np.mean(arr, axis=0)
arr_centered = arr - arr_avg
print("arr\n", arr, "\n")
print("arr_avg\n", arr_avg, "\n")
print("arr_centered\n", arr_centered, "\n")
```

```python
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = x.reshape(1, 100)
z = np.cos(x+y) + np.sin(y*x) * 5.5

plt.imshow(z, extent=[0, 10, 0, 10], origin='lower', cmap='viridis')
plt.colorbar()
```

```python
from sklearn.datasets import load_diabetes
import seaborn as sns

diabetes = load_diabetes()
x = diabetes.data.flatten() * 10
x = np.sort(x)
sns.lineplot(x)

x_min = np.min(x)
x_max = np.max(x)
x_normalized = (x - x_min) / (x_max - x_min)
sns.lineplot(x_normalized)
```

```python
arr = np.array([1, 2, 3, 4, 5])
print("arr >= 3:", arr >= 3)
print("arr % 2 == arr:", arr % 2 == arr)
print("np.count_nonzero(arr >= 3):", np.count_nonzero(arr >= 3))
print("np.sum(arr[arr >= 3]):", np.sum(arr[arr >= 3]))
print("np.any(arr % 2 == arr):", np.any(arr % 2 == arr))
print("np.all(arr > 0):", np.all(arr > 0))
```

```python
arr = np.arange(1, 1000)
s = (arr > 500) & (arr % 2 == 0)
print("np.sum( (arr > 500) & (arr % 2 == 0) ):", np.sum(s))
print("np.sum( arr[(arr > 500) & (arr % 2 == 0)] ):", np.sum(arr[s]))
```

```python
arr = np.random.randint(1, 100, size=(1000, 1000))
print("number of values over 50:", np.sum(arr > 50))
print("sum of values over 50:", np.sum(arr[arr > 50]))
```
